```python
import openpyxl
import pandas as pd
from openpyxl.styles import PatternFill

file_list = ["2022_年間売上表.xlsx", "2023_年間売上表.xlsx"]

df = pd.DataFrame()

for file_name in file_list:
    temp_df = pd.read_excel(file_name)
    temp_df.columns = ["売上年", "商品", "金額（千円）"]
    df = pd.concat([df, temp_df], ignore_index=True)

grouped = df.groupby(["商品","売上年"])["金額（千円）"].sum()

writer = pd.ExcelWriter("売上集計表.xlsx")

grouped.to_excel(writer, sheet_name="Sheet1", index=True)

writer.close()

wb = openpyxl.load_workbook("売上集計表.xlsx")

ws = wb["Sheet1"]

for row in ws["A1:C1"]:
    for cell in row:
        cell.fill = PatternFill(patternType="solid", fgColor="F2F2F2")

wb.save("売上集計表.xlsx")

# sheet = writer.sheets['Sheet1']

# for row in sheet["A1:C1"]:
#     for cell in row:
#         cell.fill = PatternFill(patternType="solid", fgColor="F2F2F2")

# writer.close()
```
